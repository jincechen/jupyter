# Econometrics Demo: LLN, CLT, and OLS Consistency

This notebook demonstrates:

1. Law of Large Numbers
2. Central Limit Theorem
3. OLS estimator via matrix algebra
4. Monte Carlo simulation of consistency

Designed for students in math/economics.

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)
```

## 1. Law of Large Numbers

We simulate sample means of i.i.d. normal variables.

```python
n = 10000
x = np.random.normal(0, 1, n)

running_mean = np.cumsum(x) / np.arange(1, n+1)

plt.plot(running_mean)
plt.axhline(0, linestyle='--')
plt.title("LLN: Convergence of Sample Mean")
plt.show()
```

## 2. Central Limit Theorem

We simulate the distribution of sample means.

```python
sample_size = 50
num_sim = 10000

means = [np.mean(np.random.exponential(1, sample_size)) for _ in range(num_sim)]

plt.hist(means, bins=50, density=True)
plt.title("CLT: Distribution of Sample Means")
plt.show()
```

## 3. OLS via Matrix Algebra

We simulate a linear model:

$$
y = X\beta + \varepsilon
$$

and compute:

$$
\hat{\beta} = (X^T X)^{-1} X^T y
$$

```python
n = 100
X = np.column_stack((np.ones(n), np.random.normal(size=n)))

beta_true = np.array([1, 2])
eps = np.random.normal(0, 1, n)

y = X @ beta_true + eps

beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y

beta_hat
```

## 4. Monte Carlo: Consistency of OLS

We increase sample size and observe convergence of estimates.

```python
def simulate_beta(n, beta_true):
    X = np.column_stack((np.ones(n), np.random.normal(size=n)))
    eps = np.random.normal(0, 1, n)
    y = X @ beta_true + eps
    beta_hat = np.linalg.inv(X.T @ X) @ X.T @ y
    return beta_hat[1]

beta_true = np.array([1, 2])
sample_sizes = [20, 50, 100, 500, 1000]

estimates = [np.mean([simulate_beta(n, beta_true) for _ in range(500)]) for n in sample_sizes]

plt.plot(sample_sizes, estimates, marker='o')
plt.axhline(2, linestyle='--')
plt.title("Consistency of OLS Slope Estimate")
plt.xlabel("Sample Size")
plt.ylabel("Average Estimate")
plt.show()
```
